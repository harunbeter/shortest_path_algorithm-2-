```python
!pip install networkx matplotlib

import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()
G.add_weighted_edges_from([
    ("A", "B", 6),
    ("A", "D", 1),
    ("B", "C", 5),
    ("B", "E", 2),
    ("C", "F", 8),
    ("D", "B", 2),
    ("D", "E", 6),
    ("E", "F", 3)
])

start_node = "A"
end_node = "F"

# En kısa yolu ve uzunluğunu buluyorum
shortest_path = nx.shortest_path(G, source=start_node, target=end_node, weight='weight')
shortest_path_length = nx.shortest_path_length(G, source=start_node, target=end_node, weight='weight')

# En kısa yolun ağırlığını hesaplıyorum
shortest_path_weight = 0
for i in range(len(shortest_path) - 1):
    shortest_path_weight += G[shortest_path[i]][shortest_path[i+1]]['weight']

print(f"En kısa yol: {shortest_path}")
print(f"En kısa yolun uzunluğu (düğüm sayısı olarak): {len(shortest_path)}")
print(f"En kısa yolun ağırlığı: {shortest_path_weight}")

pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_size=800, node_color='lightblue', font_weight='bold', font_size=7, font_color='black', font_family='Arial')
nx.draw_networkx_edge_labels(G, pos, edge_labels=nx.get_edge_attributes(G, 'weight'))
plt.show()
```
